## Creative writing

```python
# Needed in Jupyter environment See: https://ai.pydantic.dev/troubleshooting/ 
import nest_asyncio
nest_asyncio.apply()
def zero_shot(model_id, prompt: str) -> str:
    from pydantic_ai import Agent
    agent = Agent(model_id)
    result = agent.run_sync(prompt)
    return (result.data)
```

```python
essay = zero_shot(CLAUDE,
         """
Write a 4 paragraph essay that ends with the following 4 sentences respectively.
1. To be or not to be, that is the question.
2. Take me to your leader.
3. It is a truth universally acknowledged, that a single man in possession of a good fortune, must be in want of a wife.
4. The only thing we have to fear is fear itself.
Each paragraph should be 3-4 sentences.

Think step-by-step.
         """)

print(essay)
```
